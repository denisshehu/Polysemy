```python
from visualization_generation.plots import *
from models.point_cloud import *
from methods.intrinsic_dimension_estimators.danco import estimate
```

```python
n = 10000
intrinsic_dimension = 3
r = 1
seed = 1
```

```python
points = sample_from_ball(n, intrinsic_dimension, r, seed)
```

```python
plot_3d(points, True)
```

```python
point_cloud = PointCloud()
point_cloud.random_constructor(points, 3, seed=1)
```

```python
estimate(point_cloud, 10, 20)
```

```python
data = point_cloud.get_query_points()
```

```python
c = [query.intrinsic_dimension for query in point_cloud.queries]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(data[:, 0], data[:, 1], data[:, 2], c=c)
fig.colorbar(scatter)
plt.gca().set_aspect('equal')
ax.view_init(elev=30, azim=30)
plt.show()
```
