```python
from visualization_generation.plots import *
from models.point_cloud import *
from methods.euclidicity2 import calculate as old
from methods.euclidicity3 import calculate as new
from methods.intrinsic_dimension_estimators.danco import estimate
```

```python
neighborhood_size = 100
n_steps = 5
```

```python
n1, n2 = 10000, 10000
intrinsic_dimension1, intrinsic_dimension2 = 3, 3
r1, r2 = 1, 1
seed1, seed2 = 1, 1
spatial_dimension = 3
```

```python
points = sample_from_intersecting_spheres(n1, intrinsic_dimension1, r1, n2, intrinsic_dimension2, r2, spatial_dimension, seed1, seed2)
```

```python
plot_3d(points, True)
```

```python
point_cloud = PointCloud()
point_cloud.random_constructor(points, 1000)
```

```python
for query in point_cloud.queries:
    query.intrinsic_dimension = 2
```

```python
data = point_cloud.get_query_points()
c = [query.intrinsic_dimension for query in point_cloud.queries]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(data[:, 0], data[:, 1], data[:, 2], c=c)
fig.colorbar(scatter)
plt.gca().set_aspect('equal')
ax.view_init(elev=None, azim=None)
plt.show()
```
