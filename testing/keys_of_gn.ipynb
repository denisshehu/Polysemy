```python
embeddings = load_bin(normalized_gn_embeddings_path)
```

```python
keys = embeddings.index_to_key
```

```python
grouped_keys = dict()
for key in keys:
    lowercase = str.lower(key)
    if lowercase in grouped_keys.keys():
        grouped_keys[lowercase].append(key)
    else:
        grouped_keys[lowercase] = [key]
```

```python
{key: value for key, value in grouped_keys.items() if len(value) == 1}
```

```python
variants = grouped_keys['the']
variants
```

```python
for key in variants:
    most_similar = embeddings.similar_by_vector(vector=embeddings[key], topn=10)
    print(key, [key for key, similarity in most_similar])
```
