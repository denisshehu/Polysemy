```python
n1, n2 = 1000, 1000
intrinsic_dimension1, intrinsic_dimension2 = 2, 2
r1, r2 = 1, 1
spatial_dimension = 5
proportion = 0.8
```

```python
sample = sample_from_intersecting_spheres(n1, intrinsic_dimension1, r1, n2, intrinsic_dimension2, r2, spatial_dimension, proportion)
```

```python
plot_3d(sample, True, 0, -90)
```

```python
limit = r1 * (1 - proportion) / 4
```

```python
plot_3d(np.array([p for p in sample if -limit <= p[0] <= limit]), True, 0, 0)
```
